```python
import statsmodels.api as sm
```

```python
exog = train[['No Of Withdrawals','No Of XYZ Card Withdrawals','No Of Other Card Withdrawals','Weekday','Festival Religion']]
model= sm.tsa.VARMAX(train[['Total amount Withdrawn','Amount withdrawn XYZ Card']], order=(5,0), trend='nc', exog=exog)
model_result = model.fit(maxiter=1000, disp=False)
```
